### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import random
import time
import copy
import heapq
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Функция генерации матрицы
def generate_matrix(m, n, min_limit, max_limit):
    return [[random.randint(min_limit, max_limit) for _ in range(n)] for _ in range(m)]
```

```python
def selection_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    arr = arr[:]
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def shell_sort(arr):
    arr = arr[:]
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2
    return arr

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


def tournament_sort(arr):
    arr = arr[:]
    heapq.heapify(arr)
    sorted_arr = [heapq.heappop(arr) for _ in range(len(arr))]
    return sorted_arr
```

```python
# Получение параметров от пользователя
m = int(input("Введите количество строк: "))
n = int(input("Введите количество столбцов: "))
min_limit = int(input("Введите минимальное значение: "))
max_limit = int(input("Введите максимальное значение: "))

# Генерация матрицы
matrix = generate_matrix(m, n, min_limit, max_limit)
print("\nИсходная матрица:")
for row in matrix:
    print(row)
```

```python
# Копирование матрицы для тестов
matrix_copy1 = copy.deepcopy(matrix)
matrix_copy2 = copy.deepcopy(matrix)
matrix_copy3 = copy.deepcopy(matrix)
matrix_copy4 = copy.deepcopy(matrix)
matrix_copy5 = copy.deepcopy(matrix)
```

```python
# Замер времени для каждого алгоритма сортировки
def measure_sorting_time(sort_func, matrix_copy):
    start_time = time.time()
    sorted_matrix = [sort_func(row) for row in matrix_copy]
    end_time = time.time()
    return end_time - start_time

# Измеряем время для всех сортировок
selection_time = measure_sorting_time(selection_sort, matrix_copy1)
insertion_time = measure_sorting_time(insertion_sort, matrix_copy2)
bubble_time = measure_sorting_time(bubble_sort, matrix_copy3)
shell_time = measure_sorting_time(shell_sort, matrix_copy4)
quick_time = measure_sorting_time(quick_sort, matrix_copy5)
tournament_time = measure_sorting_time(tournament_sort, matrix_copy1)
```

```python

# Построение графика для времени сортировок
def plot_sorting_times():
    sorting_algorithms = ["Selection Sort", "Insertion Sort", "Bubble Sort", "Shell Sort", "Quick Sort", "Tournament Sort"]
    times = [selection_time, insertion_time, bubble_time, shell_time, quick_time, tournament_time]

    plt.figure(figsize=(10, 6))
    plt.bar(sorting_algorithms, times, color='skyblue')
    plt.xlabel('Алгоритмы сортировки')
    plt.ylabel('Время выполнения (сек.)')
    plt.title('Время выполнения различных алгоритмов сортировки')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Визуализируем график
plot_sorting_times()

# Функция для генерации случайной матрицы
def generate_matrix(m=3, n=3, min_limit=0, max_limit=10):
    return [[random.randint(min_limit, max_limit) for _ in range(n)] for _ in range(m)]

# Тепловая карта матрицы с измененными параметрами отображения
def plot_heatmap(matrix):
    plt.figure(figsize=(12, 10))  # Размер графика
    sns.heatmap(matrix, annot=True, fmt="d", cmap="YlGnBu", cbar=True, 
                annot_kws={"size": 6}, linewidths=0.5, linecolor='black', square=True)  # улучшено отображение
    plt.title("Тепловая карта матрицы")  # Заголовок
    plt.xlabel("Столбцы")  # Подпись оси X
    plt.ylabel("Строки")  # Подпись оси Y
    plt.show()  # Отображение графика

# Генерация случайной матрицы 20x20 для теста (если хотите 100x100, поменяйте параметры)
matrix = generate_matrix(20, 20, 0, 100)

# Визуализируем тепловую карту
plot_heatmap(matrix)
```
